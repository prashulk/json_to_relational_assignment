### Read Brands

```python
import pandas as pd
import json
from pandas import json_normalize

with open('brands.json', 'r') as file:
    json_data = file.readlines()

parsed_data = [json.loads(line) for line in json_data]

df = json_normalize(parsed_data)

df.head()
```

```python
df.info()
```

### Nulls percentage

```python
import pandas as pd

null_counts = df.isnull().sum()
total_rows = len(df)
null_percentages = round((null_counts / total_rows) * 100,2)

print("Null Counts:")
print(null_counts)
print("\nNull Percentages:")
print(null_percentages)
```

### Duplicate verification based on uuid

```python
duplicate_count = df.duplicated(subset=['_id.$oid']).sum()

print(duplicate_count)
```

```python
df.tail(10)
```

### Replacing Blanks with None for consistency

```python
import numpy as np
df = df.replace({np.NaN: None})

for col in df.columns:
    df[col] = df[col].replace('', np.nan)
```

### Renaming columns and sequencing

```python
column_mapping = {
    '_id.$oid': 'brand_uuid',
    'cpg.$id.$oid': 'cpg_id',
    'cpg.$ref': 'cpg_ref',
}

df_mapped = df.rename(columns=column_mapping)

columns_sequence = ['brand_uuid', 'barcode', 'brandCode', 'category', 'categoryCode', 'cpg_id', 'cpg_ref', 'topBrand', 'name']
df_mapped_reordered = df_mapped[columns_sequence]

df_mapped_reordered.tail(10)
```

### Connect to database and create table

```python
import psycopg2
from sqlalchemy import create_engine, Boolean, VARCHAR, TIMESTAMP, BIGINT

db_config = {
    'user': 'postgres',
    'password': '9901',
    'host': 'localhost',
    'port': '5433',
    'database': 'postgres'
}

conn = psycopg2.connect(**db_config)

cursor = conn.cursor()

engine = create_engine('postgresql+psycopg2://{user}:{password}@{host}:{port}/{database}'.format(**db_config))

dtype_mapping = {

    'brand_uuid': VARCHAR(350),
    'barcode': VARCHAR(30),
    'brandCode': VARCHAR(500),
    'category': VARCHAR(500),
    'categoryCode': VARCHAR(350),
    'cpg_id': VARCHAR(350),
    'cpg_ref': VARCHAR(100),
    'topBrand': Boolean,
    'name': VARCHAR(350)
}

df_mapped_reordered.to_sql('brand', engine, if_exists='append', index=False, dtype=dtype_mapping)

conn.commit()
conn.close()
```
