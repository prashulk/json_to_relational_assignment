### Break the input JSON to two dataframes and append receipt_id to each

```python
import pandas as pd
import json

with open('receipts.json', 'r') as file:
    json_data = file.readlines()

parsed_data = [json.loads(line) for line in json_data]

receipts_fact_df = pd.json_normalize(parsed_data)

items_list = []
for entry in parsed_data:
    receipt_id = entry['_id']['$oid']
    if 'rewardsReceiptItemList' in entry:
        items = entry['rewardsReceiptItemList']
        for item in items:
            item['receipt_id'] = receipt_id  # Add receiptId to each item
            items_list.append(item)

rewards_receipt_items_df = pd.DataFrame(items_list)

receipts_fact_df.drop(columns=['rewardsReceiptItemList'], inplace=True)
```

### Creation of Integration_ID as primary key based on partneritemID and receiptID

```python
# Create a new column 'integration_id' as a combination of partner item ID and receipt ID
rewards_receipt_items_df['integration_id'] = rewards_receipt_items_df['partnerItemId'].astype(str) + '~' + rewards_receipt_items_df['receipt_id']

rewards_receipt_items_df.head()
```

### Date conversion

```python
from datetime import datetime

def epoch_to_datetime(epoch_time):
    if pd.isnull(epoch_time):
        return None
    else:
        return datetime.fromtimestamp(epoch_time / 1000)

receipts_fact_df['createDate'] = receipts_fact_df['createDate'].apply(epoch_to_datetime)
receipts_fact_df['dateScanned'] = receipts_fact_df['dateScanned'].apply(epoch_to_datetime)
receipts_fact_df['finishedDate'] = receipts_fact_df['finishedDate'].apply(epoch_to_datetime)
receipts_fact_df['modifyDate'] = receipts_fact_df['modifyDate'].apply(epoch_to_datetime)
receipts_fact_df['pointsAwardedDate'] = receipts_fact_df['pointsAwardedDate'].apply(epoch_to_datetime)
receipts_fact_df['purchaseDate'] = receipts_fact_df['purchaseDate'].apply(epoch_to_datetime)
```
