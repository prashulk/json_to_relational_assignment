```python
import pandas as pd
import json
from pandas import json_normalize

with open('users.json', 'r') as file:
    json_data = file.readlines()

parsed_data = [json.loads(line) for line in json_data]

df = json_normalize(parsed_data)

df.head()
```

```python
df.info()
```

### Get Null percentage counts

```python
import pandas as pd

null_counts = df.isnull().sum()
total_rows = len(df)
null_percentages = round((null_counts / total_rows) * 100,2)

print("Null Counts:")
print(null_counts)
print("\nNull Percentages:")
print(null_percentages)
```

### Duplication in User Id's

```python
unique_count = df['_id.$oid'].nunique()

print("Count of Unique Values based on _id.$oid:")
print(unique_count)
```

```python
df.tail(10)
```

### Handling NaN and blank string values

```python
import numpy as np
df = df.replace({np.NaN: None})

for col in df.columns:
    df[col] = df[col].replace('', np.nan)
```

```python
df.tail(10)
```

### Removing Duplicates based on Primary Key

```python
df_no_duplicates = df.drop_duplicates(subset=['_id.$oid'])
```

```python
df_no_duplicates.info()
```

### Column Renaming

```python
column_mapping = {
    'active': 'active',
    'role': 'role',
    'signUpSource': 'signUpSource',
    'state': 'state',
    '_id.$oid': 'user_id',
    'createdDate.$date': 'createdDate',
    'lastLogin.$date': 'lastLogin'
}

df_mapped = df_no_duplicates.rename(columns=column_mapping)

columns_sequence = ['user_id', 'active', 'role', 'signUpSource', 'state', 'createdDate', 'lastLogin']
df_mapped_reordered = df_mapped[columns_sequence]

df_mapped_reordered.tail(10)
```

### Convert Epoch to timestamp

```python
from datetime import datetime

def epoch_to_datetime(epoch_time):
    if pd.isnull(epoch_time):
        return None
    else:
        return datetime.fromtimestamp(epoch_time / 1000)

df_mapped_reordered['createdDate'] = df_mapped_reordered['createdDate'].apply(epoch_to_datetime)
df_mapped_reordered['lastLogin'] = df_mapped_reordered['lastLogin'].apply(epoch_to_datetime)

df_mapped_reordered.tail(10)
```

```python
df_mapped_reordered.info()
```

### Inserting into database

```python
import psycopg2
from sqlalchemy import create_engine, Boolean, VARCHAR, TIMESTAMP

db_config = {
    'user': 'postgres',
    'password': '9901',
    'host': 'localhost',
    'port': '5433',
    'database': 'postgres'
}

conn = psycopg2.connect(**db_config)

cursor = conn.cursor()

engine = create_engine('postgresql+psycopg2://{user}:{password}@{host}:{port}/{database}'.format(**db_config))

dtype_mapping = {
    'active': Boolean,
    'role': VARCHAR(350),
    'signUpSource': VARCHAR(700),
    'state': VARCHAR(50),
    'user_id': VARCHAR(700),
    'createdDate': TIMESTAMP,
    'lastLogin': TIMESTAMP
}

df_mapped_reordered.to_sql('users', engine, if_exists='append', index=False, dtype=dtype_mapping)

conn.commit()
conn.close()
```
